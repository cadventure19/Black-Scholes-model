```python
import pandas as pd
import numpy as np
import math
import scipy.stats as si
```

```python
datareader=pd.read_csv('netflix2.csv')
df=datareader
```

```python

df.drop(['Open', 'High','Low','Adj Close'], axis=1, inplace=True)
del df['Volume']
```

```python

df['Week'] = np.arange(len(df))
```

```python

r=0.05848
K=370
sigma=.0994
```

```python
df['T']=(df['Week']*(7)/(254))
```

```python
df['d1']=((np.log(df['Close'])/K)+(r+(.5*sigma*sigma))*(df['T']))/(sigma*np.sqrt(df['T']))
```

```python
df['d1']=((np.log((df['Close'])/K)+(r+(.5*sigma*sigma))*(df['T'])))/((sigma*np.sqrt(df['T'])))
```

```python
df['d2']=df['d1']-sigma*np.sqrt(df['T'])
```

```python
df['Call_price']=(df['Close']*si.norm.cdf(df['d1'],0.0,1.0))-(K*np.exp(-r*df['T'])*si.norm.cdf(df['d2'],0.0,1.0))
```

```python
df['delta']=si.norm.cdf(df['d1'],0.0,1.0)
```

```python
df['Number of stocks']=np.ceil(df['delta']*1000)
```

```python
df['Received']=df['Call_price']*1000
```

```python
df['Shares Bought']=df['Number of stocks']*((df['Close']).shift(1))
```

```python
df['Borrowing']=df['Shares Bought']-df['Received']
```

```python

df.loc[2:,'Borrowing'] =df['Borrowing'].shift(1)*(np.exp(r*df['T']))+(df['Number of stocks']-df['Number of stocks'].shift(1))*df['Call_price']
```

```python
df['Received'][2:]=0
df['Shares Bought'][2:]=0
```
